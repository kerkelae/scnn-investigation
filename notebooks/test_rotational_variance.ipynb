```python
import torch
```

```python
for i, (fname, model_name) in enumerate(
    zip(
        [
            "../rotation_scnn_preds.pt",
            "../rotation_scnn_preds_rot.pt",
            "../rotation_smt_preds.pt",
            "../rotation_mlp_preds.pt",
            "../rotation_mlp_preds_rot.pt",
        ],
        [
            "sCNN",
            "sCNN*",
            "SMT",
            "MLP",
            "MLP*",
        ],
    )
):
    preds = torch.load(fname).reshape((1000, -1, 47))

    print(model_name)
    print(f"d: {torch.std(preds[:, :, -2], dim=-1).mean().item() * 1e3}")
    print(f"f: {torch.std(preds[:, :, -1], dim=-1).mean().item() * 1e3}")
    print("")
```
